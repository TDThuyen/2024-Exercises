**KNN**

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

iris_X, iris_y = datasets.load_iris(return_X_y = True)

X_train, X_test, y_train, y_test = train_test_split(
    iris_X,
    iris_y,
    test_size = 0.2,
    random_state = 42
)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

knn_classifier = KNeighborsClassifier(n_neighbors= 5)
knn_classifier.fit(X_train, y_train)

y_pred = knn_classifier.predict(X_test)
accuracy_score(y_test, y_pred)


```

**K-mean**

```python
import numpy as np

# dataset
data = np.array([
    [2.0, 3.0, 1.5],
    [3.0, 3.5, 2.0],
    [3.5, 3.0, 2.5],
    [8.0, 8.0, 7.5],
    [8.5, 8.5, 8.0],
    [9.0, 8.0, 8.5],
    [1.0, 2.0, 1.0],
    [1.5, 2.5, 1.5]
])

data.shape
```

```python
centroids = np.array([[2.0, 3.0, 1.5], [1.0, 2.0, 1.0], [3.0, 3.5, 2.0]])
centroids.shape
```

```python
data = data[:, np.newaxis, :]
data
```

```python
iter = 10
for i in range(iter):
  # compute distance
  distances = np.sqrt(np.sum(np.square(data - centroids), axis = 2))

  print(distances)

  #labels:
  labels = np.argmin(distances, axis = 1)
  print(labels)

  # X = [data[labels == j].mean(axis = 0) for j in range(2)]
  # print(X)
  # break
  #updates centroids:
  new_centroids = np.array([data[labels == j].mean(axis = 0) for j in range(3)]).reshape(3, 3)
  print(new_centroids)
  print(new_centroids.shape)


  #check:
  if np.all(centroids == new_centroids):
    break
  centroids = new_centroids

```
